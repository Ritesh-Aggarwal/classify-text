```python
import csv
import nltk
import re
from nltk.corpus import wordnet
# nltk.download('wordnet')
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
# nltk.download('omw-1.4')
```

```python
camelCase = re.compile('[a-z]+((\d)|([A-Z0-9][a-z0-9]+))*([A-Z])?')
snake_case = re.compile('[a-z]+(_[a-z]+)*')
key_dict = ['if','while','for','then','echo','def','len','range','exit'
            ,'function','else','elif','void','return','int','Integer','console.log','var',
            'float','double','string','String','const','try','catch','const','#include','private','public','static','async']

bash_commands = ['ls','cd','mv','mkdir','mv','rmdir','rm','touch','clear','cp']

def count_bashcommads(setence):
    words = nltk.word_tokenize(sentence)
    count_bashcommands =0
    for word in words:
        if word in bash_commands :
            count_bashcommands = count_bashcommands + 1
    print(count_bashcommands)
    return count_bashcommands

def count_keyword(sentence):
    words = nltk.word_tokenize(sentence)
    count_keyword =0
    for word in words:
        if word in key_dict :
            count_keyword = count_keyword + 1
    print(count_keyword)
    return count_keyword

def dictionary_nondictionary(sentence):
    words = nltk.word_tokenize(sentence)
    count_nondictionary = 0
    for word in words:
        if wordnet.synsets(word):
            continue
        else :
            count_nondictionary = count_nondictionary +1
    print(count_nondictionary/len(words))
    return count_nondictionary/len(words)

def count_plus(sentence):
    words = nltk.word_tokenize(sentence)
    count_plus = 0
    for word in words:
        for i in range(len(word)):
            if word[i] =='+':
                count_plus = count_plus+1
            else :
                continue
    print(count_plus)
    return count_plus

def count_minus(sentence):
    words = nltk.word_tokenize(sentence)
    count_minus = 0
    for word in words:
        for i in range(len(word)):
            if word[i] =='-':
                count_minus = count_minus+1
            else :
                continue
    print(count_minus)
    return count_minus

def count_slash(sentence):
    words = nltk.word_tokenize(sentence)
    count_slash = 0
    for word in words:
        for i in range(len(word)):
            if word[i] =='/' or word[i]=='\\' :
                count_slash = count_slash + 1
            else :
                continue
    print(count_slash)
    return count_slash

def count_equal(sentence):
    words = nltk.word_tokenize(sentence)
    count_equal = 0
    for word in words:
        for i in range(len(word)):
            if word[i] =='=':
                count_equal = count_equal+1
            else :
                continue
    print(count_equal)
    return count_equal

def count_colon(sentence):
    words = nltk.word_tokenize(sentence)
    count_colon = 0
    for word in words:
        for i in range(len(word)):
            if word[i] ==':' or word[i]==';':
                count_colon = count_colon+1
            else :
                continue
    print(count_colon)
    return count_colon

def count_mul(sentence):
    words = nltk.word_tokenize(sentence)
    count_mul = 0
    for word in words:
        for i in range(len(word)):
            if word[i] =='*' :
                count_mul = count_mul+1
            else :
                continue
    print(count_mul)
    return count_mul

def count_specialchar(sentence):
    words = nltk.word_tokenize(sentence)
    count_splchar = 0
    for word in words:
        for i in range(len(word)):
            ch =word[i]
            if ch=='#' or ch=='{' or ch=='}' or ch=='[' or ch==']' or ch=='&' or ch=='%' or ch=='^' or ch=='<' or ch=='>' or ch=='_' or ch=='|' or ch=='!':
                count_splchar = count_splchar+1
            else :
                continue
    print(count_splchar)
    return count_splchar

def nontextual_word(word):
    for i in range(len(word)):
        ch = word[i]
        if ch.isdigit():
            return 1
        elif ch == '+' or ch == '-' or ch == '/' or ch=='*' or ch == '^' or ch ==':' or ch ==';' or ch =='' or ch=='{':
            return 1
        elif ch.isalpha() or ch ==',' or ch == '&' or ch =='.':
            continue
        else :
            return 1
    return 0
            

def textual_nontextual(str):
    nltk_tokens = nltk.word_tokenize(str)
    count_textual=0
    count_nontextual=0
    for i in range(len(nltk_tokens)):
        word = nltk_tokens[i]
        if nontextual_word(word):
            count_nontextual +=1
        else :
            count_textual +=1
    #print (nltk_tokens)
    print (count_nontextual/len(nltk_tokens))
    return count_nontextual/len(nltk_tokens)

def to_sentences(text):
    sentences = nltk.sent_tokenize(text)
    print("sentences are : ")
    print(sentences)
    for sentence in sentences:
        textual_nontextual(sentence)
    return sentences

def possible_camelCase(sentence):
    words = nltk.word_tokenize(sentence)
    count =0
    for word in words:
        if camelCase.match(word):
                    count=count+1
                    break
    print(count)
    return count

def underscores(sentence):
    words = nltk.word_tokenize(sentence)
    count=0
    for word in words:
        if snake_case.match(word):
                    count=count+1
                    break
    print(count)
    return count
```

```python
data.columns = ["URL", "Text", "Start-Indices", "End-Indices", "Text-Type"]
data["Text"]
```

```python
#for row in data.iterrows():

data["plus"] = data.apply(lambda row: count_plus(row["Text"]), axis=1)
data["minus"] = data.apply(lambda row: count_minus(row["Text"]), axis=1)
data["mul"] = data.apply(lambda row: count_mul(row["Text"]), axis=1)
data["underscore"] = data.apply(lambda row: underscores(row["Text"]), axis=1)
data["camelcase"] = data.apply(lambda row: possible_camelCase(row["Text"]), axis=1)
data["keyword"] = data.apply(lambda row: count_keyword(row["Text"]), axis=1)
data["dictionary-nondictionary-terms"] = data.apply(lambda row: dictionary_nondictionary(row["Text"]), axis=1)
data["textual-nontexual"] = data.apply(lambda row: textual_nontextual(row["Text"]), axis=1)

target_var = data["Text-Type"]
data_new = data.drop(["URL", "Text", "Start-Indices", "End-Indices", "Text-Type"], axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_new, target_var, test_size=0.1, random_state=1)
print("New Training set size:", X_train.shape)
print("New Test set size:", X_test.shape)
```

```python
import xgboost as xgb
from xgboost import plot_importance
from sklearn.model_selection import cross_val_score
 
model_classifier = xgb.XGBClassifier(n_estimators=1000, objective = 'multi:softmax', num_class = 5,
                                     metrics = 'mlogloss', learning_rate=0.03, max_depth = 7,
                                     min_child_weight = 0.7, gamma = 0, silent = 0, subsample = 0.8,
                                     seed = 27, colsample_bytree = 1)
model_classifier = model_classifier.fit(X_train,y_train)
plot_importance(model_classifier)
```
