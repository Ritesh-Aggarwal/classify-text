```python
import pandas as pd
data = pd.read_csv("./data.csv")
data.columns = ["URL", "Text", "Start-Indices", "End-Indices", "Text-Type"]
#data["Text-Type"] = 5
data.head()

trainDF = pd.DataFrame()
trainDF['text'] = data["Text"]
trainDF['label'] = data["Text-Type"]

import nltk
sentence_corpus=[]
for sentence in trainDF["text"]:
    sentence_corpus.append(nltk.word_tokenize(sentence))

!pip install glove-python

from glove import Corpus, Glove

# Create corpus dictionary of all the words in the training corpus
corpus = Corpus()
corpus.fit(sentence_corpus, window=10)  # 'tokens' is a list of tokenized sentences from the training corpus

# Create co-occurrence matrix of the training corpus using the Corpus() object
corpus_matrix = corpus.matrix

# Train the Glove model with word vector embeddings dimension as 300
glove = Glove(no_components=300, learning_rate=0.05)
glove.fit(corpus_matrix, epochs=30, no_threads=4, verbose=True)

# Create an embedding index dictionary which has keys as corpus words and values as their corresponding word embeddings
embedding_index = {}
for word in corpus.dictionary:
    embedding_index[word] = glove.word_vectors[glove.dictionary[word]]
```
